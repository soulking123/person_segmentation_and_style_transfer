```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("image/dog.png")
mask = cv2.imread("image/dog_mask.png")
bg = cv2.imread("image/bg.jpg")
bg = cv2.resize(bg,(512,422))
normalized_image = cv2.normalize(mask, None, 0, 1, cv2.NORM_MINMAX)
```

```python
print(image.shape)
print(mask.shape)
print(bg.shape)
print(normalized_image.shape)
```

```python
result1 = image * normalized_image
plt.imshow(cv2.cvtColor(result1,cv2.COLOR_BGR2RGB))
```

```python
inverted_mask = cv2.bitwise_not(mask)//255
result2 = bg * inverted_mask
plt.imshow(cv2.cvtColor(result2,cv2.COLOR_BGR2RGB))
```

```python
result3 = result1+result2
plt.imshow(cv2.cvtColor(result3,cv2.COLOR_BGR2RGB))
```
